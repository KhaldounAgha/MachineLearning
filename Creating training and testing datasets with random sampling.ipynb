```python
import pandas as pd
file_path = 'data\\xlsx linear regression ml project.xlsx'
file = pd.read_excel(io= file_path, sheet_name= 'fitness_trackers_project')
df = pd.DataFrame(data= file)
print(f'The data frame includes [{df.shape[0]}] rows and [{df.shape[1]}] columns.')
```

```python
## IMPORTANT: 

## "replace" parameter determines whether you want to sample with or without replacement.
#  "replace=False" means that a particular row or observation can only be selected once in the sample (ensure that each sampled row is unique and not repeated).
## "replace=True" means that a particular row can be selected multiple times in the sample.

# Note: The reason why we reset beta_df_train index after creating berta_df_test

beta_df_train = df.sample(frac= 0.8, random_state= 42, replace= False)

# Please inspect the index of the beta_df_train to understand the following steps. Use beta_df_train.head()
beta_df_test = df.drop(index= beta_df_train.index).reset_index(drop= True)
beta_df_train.reset_index(drop= True, inplace= True)

print(f'Train set shape: [{beta_df_train.shape}], Test set shape [{beta_df_test.shape}]')
```
