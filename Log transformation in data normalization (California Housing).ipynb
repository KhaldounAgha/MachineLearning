```python
import numpy as np
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
```

```python
# Load the California housing dataset
X, y = fetch_california_housing(return_X_y=True)

print(f'The California Housing dataset includes {X.shape[0]} observations on {X.shape[1]} variables') 
print(f'The California Housing dataset includes {y.shape[0]} target prices') 
```

```python
# Convert house prices to original scale
original_price = np.array(y * 100000, dtype=int)
print(f'Example of original house prices: {original_price[:5]}')
```

```python
# Plot y and original_price
plt.figure(figsize=(10, 4))

# Plot y (in units of $100,000)
plt.subplot(1, 2, 1)
plt.hist(y, bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of y (in $100,000s)")
plt.xlabel("Median house price (in $100,000s)")
plt.ylabel("Frequency")

# Plot original_price (in dollars)
plt.subplot(1, 2, 2)
plt.hist(original_price, bins=30, color='salmon', edgecolor='black')
plt.title("Distribution of original_price (in dollars)")
plt.xlabel("Median house price (in dollars)")
plt.ylabel("Frequency")

# Display the plots
plt.tight_layout()
plt.show()
```

```python
y_log_scaled = np.log1p(y) # Log(1 + x)
original_price_log_scaled = np.log1p(y)
```

```python
plt.figure(figsize=(10, 4))

# Plot y (in units of $100,000)
plt.subplot(1, 2, 1)
plt.hist(y_log_scaled, bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of y log scaled (in $100,000s)")
plt.xlabel("Median house price (in $100,000s)")
plt.ylabel("Frequency")

# Plot original_price (in dollars)
plt.subplot(1, 2, 2)
plt.hist(original_price_log_scaled, bins=30, color='salmon', edgecolor='black')
plt.title("Distribution of original_price (in dollars)")
plt.xlabel("Median house price (in dollars)")
plt.ylabel("Frequency")

# Display the plots
plt.tight_layout()
plt.show()
```
