```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
sales = [330.17, 498.82, 478.52, 451.32, 565.38, 601.42, 845.30, 619.36, 969.84, 995.62]
advertising = [20.00, 24.44, 28.89, 33.33, 37.78, 42.22, 46.67, 51.11, 55.56, 60.00]
```

```python
plt.figure(figsize=(8, 3))
plt.style.use('bmh')
plt.scatter(x=advertising, y=sales, marker='o', c='darkblue', alpha=0.9, edgecolors='black')
plt.xlabel('Advertising', fontsize=10, fontweight='bold', color='black', font='Arial')
plt.ylabel('Sales', fontsize=10, fontweight='bold', color='black', font='Arial')
plt.title('Advertising vs Sales', fontdict={'fontsize': 12, 'fontweight': 'bold', 'color': 'darkblue', 'font': 'Arial'}, loc='center')
plt.grid(visible=True, color='dimgrey', linestyle='dashed', linewidth=0.4)
plt.tick_params(axis='both', labelsize=9, color='black', labelcolor='black', length=4, width=1.2, grid_color='dimgrey', grid_linewidth=0.4, grid_alpha=0.8)
plt.show()
```
