##### **train_test_split()**

```python
from sklearn.model_selection import train_test_split
```

```python
# train_test_split(arrays: )
# *arrays : sequence of indexables with same length / shape[0]
# Allowed inputs are lists, numpy arrays, scipy-sparse matrices or pandas dataframes.
```

##### # Splitting the data into training and testing sets using [NumPy arrays]

```python
# A - Note: One way to load the Iris dataset with features (X) and target (y) as separate arrays
X = iris_X_y[0]
y = iris_X_y[1]

# B - Note: Another way to load the Iris dataset with features (X) and target (y) as separate arrays
X = iris.data
y = iris.target
```

```python
# Splitting the data into training and testing sets using NumPy arrays
X_train, X_test, y_train, y_test = train_test_split(iris_X_y[0], iris_X_y[1], test_size=0.2, random_state=42, shuffle=True) # Or use X, y
```

```python
print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
```

##### **Model Selection**

<style> .small-text { font-size: 13px; } </style> <p class="small-text"> 
A- Linear Models: 

1- Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logistic_regression_model = LogisticRegression()
```

```python
logistic_regression_model.fit(X=X_train, y=y_train)
```

```python
y_pred = logistic_regression_model.predict(X=X_test)
```

```python
prediction_df_columns = list(iris['feature_names'])
prediction_df_columns.extend(['Target', 'Predictions'])
predictions_np_array = np.concatenate((X_test, y_test.reshape(-1, 1), y_pred.reshape(-1, 1)), axis=1)
predictions_df = pd.DataFrame(data=predictions_np_array, columns=prediction_df_columns)
predictions_df.head(10)
```

```python
index_list = predictions_df.loc[predictions_df['Target'] != predictions_df['Predictions']].index
print(f'Number of not corrected predictions: {len(index_list)}')
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report

mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
print(f'Mean Squared Error: {mse}%')
```

```python
accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
print("Accuracy:", accuracy)
```

```python
print("\nClassification Report:\n", classification_report(y_true=y_test, y_pred=y_pred))
```

```python
# Optional: Predict class probabilities
y_probs = logistic_regression_model.predict_proba(X_test)
print("\nPredicted Probabilities:\n", y_probs)
```

```python
predictions_probability_array = np.concatenate((predictions_np_array, np.round(y_probs, 2)), axis=1)
prediction_df_columns.extend([f'Class_{i}_prob' for i in range(y_probs.shape[1])])
predictions_probability_df = pd.DataFrame(data=predictions_probability_array, columns=prediction_df_columns)
```
