```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# Generate data 
X = np.random.randint(1000, 5000, size=1000)
print(f'Shape of the vector array: {X.shape}')

# Min-Max Scaling (Normalization) (Range [0, 1])
min = np.min(X)
max = np.max(X)
X_MinMaxScaled = (X - min) / (max - min)

# Plot the histograms
plt.figure(figsize=(10, 3), facecolor='lightgrey')

plt.subplot(1, 2, 1)
plt.hist(x=X, bins=25, edgecolor='black', alpha=0.7, color='salmon')
plt.title('Original Feature')

plt.subplot(1, 2, 2)
plt.hist(x=X_MinMaxScaled, bins=25, edgecolor='black', alpha=0.7, color='skyblue')
plt.title('Min-Max Scaled Feature')

plt.tight_layout()
plt.show()
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Constructing an ndarray with features of different scales
X_train = np.array([
    np.random.randint(0, 10, 100),
    np.random.randint(100, 500, 100),
    np.random.randint(-50, 50, 100), 
    np.random.normal(loc=20, scale=15, size=100)
]).T
```

```python
# Min-Max Scaling (Normalization) (Range [0, 1])
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(X_train)
X_train_MinMaxScaled = min_max_scaler.transform(X_train)
```

```python
# Plotting
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(18, 6), facecolor='lightgrey') 
fig.suptitle('Feature Distributions Before and After Min-max Scaling', size=12) 

feature_names = ['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4']  

for i in range(4):
    axes[0, i].hist(X_train[:, i], bins=25, edgecolor='black', color='royalblue', alpha=0.8)
    axes[0, i].set_title(f'{feature_names[i]} (Original)') 
    axes[0, i].set_ylabel("Frequency")

    axes[1, i].hist(X_train_MinMaxScaled[:, i], bins=25, edgecolor='black', color='salmon', alpha=0.8)
    axes[1, i].set_title(f'{feature_names[i]} (MinMaxScaled)')
    axes[1, i].set_ylabel("Frequency")

plt.tight_layout() 
plt.show()
```
