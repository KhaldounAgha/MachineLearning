**Methods to Transform Skewed Data**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import FunctionTransformer
```

```python
# Transforming skewed data (making it more normally distributed)
# is crucial because "many" machine learning algorithms assume that the data follows a normal distribution. 
# Skewness can lead to biased estimates and affect the performance of your models.
```

**Log Transformation (Manual)**

```python
# Log Transformation (Manual)
# Suitable for highly skewed positive data.
# Can be applied using NumPy before feeding into Scikit-learn models.
```

```python
# Generate data (highly skewed positive data (exponential distribution))
np.random.seed(42)
skewed_data = np.random.exponential(scale=2, size=(1000, 1))  # Scale controls the spread
print(f'Shape of the skewed data: {skewed_data.shape}')
```

```python
# Apply log transformation
transformed_data = np.log1p(skewed_data) # Applies a log transformation (log(1 + x))
print(f'Shape of the transformed data: {transformed_data.shape}')
```

```python
# Plotting
fig, axes = plt.subplots(nrows=1, ncols=2) #  fig, (ax1, ax2)
fig.set_figwidth(10)
fig.set_figheight(3)
fig.set_facecolor('lightgrey')
fig.suptitle('Comparison of Original Skewed Feature and the Log Transformed Feature', fontsize=11)

title_list = ['Feature 1 (Skewed)', 'Feature 2 (Log Transformed)']

axes[0].hist(skewed_data, bins=30, color='navy', edgecolor='black', alpha=0.6)
axes[0].set_title(title_list[0], fontsize=10)

axes[1].hist(transformed_data, bins=30, color='olivedrab', edgecolor='black', alpha=0.6)
axes[1].set_title(title_list[1], fontsize=10)

plt.tight_layout()
plt.show()
```

**Log Transformation (sklearn)**

```python
log_transformer = FunctionTransformer(func=np.log1p, inverse_func=np.expm1, validate=True)
log_transformed_data = log_transformer.fit_transform(skewed_data)
```

```python
# Plotting
fig, axes = plt.subplots(nrows=1, ncols=2) #  fig, (ax1, ax2)
fig.set_figwidth(10)
fig.set_figheight(3)
fig.set_facecolor('lightgrey')
fig.suptitle('Comparison of Original Skewed Feature and the Log Transformed Feature', fontsize=11)

title_list = ['Feature 1 (Skewed)', 'Feature 2 (Log Transformed - sklearn)']

axes[0].hist(skewed_data, bins=30, color='gold', edgecolor='black', alpha=0.6)
axes[0].set_title(title_list[0], fontsize=10)

axes[1].hist(transformed_data, bins=30, color='steelblue', edgecolor='black', alpha=0.6)
axes[1].set_title(title_list[1], fontsize=10)

plt.tight_layout()
plt.show()
```
