```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MaxAbsScaler
```

```python
# Generate Data
np.random.seed(42)
X = np.array(np.random.randint(low=-40, high=80, size=900))
print(f'Shape of the vector array: {X.shape}')

# MaxAbs Scaling (Normalization) (Range [-1, 1])
X_MaxAbsScaled = X / np.max(np.abs(X))

# Plotting
plt.figure(figsize=(10, 3), facecolor='lightgrey')
plt.suptitle('Comparison of Original and MaxAbs Scaled Features', fontsize=11)

plt.subplot(1, 2, 1)
plt.hist(X, bins=25, color='navy', edgecolor='black', alpha=0.7)
plt.title('Original')

plt.subplot(1, 2, 2)
plt.hist(X_MaxAbsScaled, bins=25, color='salmon', edgecolor='black', alpha=0.8)
plt.title('MaxAbs Scaled')

plt.tight_layout()
plt.show()
```

```python
# Generate the dataset
X_train = np.array([
    np.random.randint(-300, 600, 900),   # Large range of integers
    np.random.randint(-100, 70, 900),    # Smaller range of integers
    np.random.uniform(-2.5, 3.0, 900),   # Floating-point values with small range
    np.random.uniform(-1.2, 1.0, 900)    # Even smaller floating-point range
]).T

print(X_train.shape)  # Should be (900, 4)
```

```python
# MaxAbs Scaling (Normalization) (Range [-1, 1])
max_abs_scaler = MaxAbsScaler()
max_abs_scaler.fit(X_train)
X_train_MaxAbsScaled = max_abs_scaler.transform(X_train)
```

```python
# Plotting
fig, axes = plt.subplots(nrows=2, ncols=4)
fig.suptitle('Feature Distributions Before and After MaxAbs scaling', size=12)
fig.set_facecolor('lightgrey')
fig.set_figwidth(18)
fig.set_figheight(6)

feature_names = ['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4']

for i in range(4):
    axes[0, i].hist(X_train[:, i], bins=25, edgecolor='black', color='royalblue', alpha=0.8)
    axes[0, i].set_title(f'{feature_names[i]} (Original)')
    axes[0, i].set_ylabel('Frequency')

    axes[1, i].hist(X_train_MaxAbsScaled[:, i], bins=25, edgecolor='black', color='salmon', alpha=0.8)
    axes[1, i].set_title(f'{feature_names[i]} (MaxAbs Scaled)')
    axes[1, i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
