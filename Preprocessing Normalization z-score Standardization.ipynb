```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import random
```

```python
## Data Normalization [# Standardization (Z-score Normalization) #] Mathematical formula and Standardization in Scikit-Learn (sklearn)
```

```python
# Z-score standardization (z-score normalization) transforms the data to have a mean of 0 and a standard deviation of 1,
# but it does not change the underlying distribution of the data. 
# If the original data is skewed, the standardized data will retain the same skewness.
# Z-score standardization shifts and scales the data but does not alter its shape. Skewness, kurtosis, and other higher-order moments remain unchanged.
# If you want to make skewed data more normally distributed, you would need to apply a transformation (e.g., log transformation, square root)
# If normality is required for your modeling, consider using a transformation method in addition to or instead of z-score standardization.
```

```python
# Standardization (Z-score Normalization) Mathematical formula
```

```python
# Generate data 
X = np.random.randint(1000, 5000, size=1000)
print(f'Shape of the vector array: {X.shape}')
# Standardization
mean = np.mean(X)
std = np.std(X)  # ddof=0 for population std
X_Standardized = (X - mean) / std

# Plot the histograms
plt.figure(figsize=(10, 3), facecolor='lightgrey')

plt.subplot(1, 2, 1)
plt.hist(X, bins=25, edgecolor="black", alpha=0.8, color='salmon')
plt.title('Original Feature')

plt.subplot(1, 2, 2)
plt.hist(X_Standardized, bins=25, edgecolor="black", alpha=0.7, color='navy')
plt.title('Standardized Feature')

plt.tight_layout()
plt.show()
```

```python
# Set random seed for reproducibility
random.seed(42)

# Generate synthetic data with features of different scales
X_train = np.array([
    random.choices(range(1, 100), k=1000),  # Feature 1: Small range
    random.choices(range(1000, 5000), k=1000),  # Feature 2: Large range
    [random.gauss(mu=50, sigma=5) for _ in range(1000)]  # Feature 3: Gaussian distribution
]).T

# Standardization (Z-score Normalization)
standard_scaler = StandardScaler()
X_train_standardized = standard_scaler.fit_transform(X_train)
```

```python
# Plotting
fig, axes = plt.subplots(2, 3, figsize=(12, 4), facecolor='lightgrey')
fig.suptitle('Feature Distributions Before and After Standardization', size=12)

# Plot original features
titles = ['Feature 1 (Original)', 'Feature 2 (Original)', 'Feature 3 (Original)']
for i, ax in enumerate(axes[0]):
    ax.hist(X_train[:, i], bins=25, color='skyblue', alpha=0.8, edgecolor='black')
    ax.set_title(titles[i])

# Plot standardized features
titles_std = ['Feature 1 (Standardized)', 'Feature 2 (Standardized)', 'Feature 3 (Standardized)']
for i, ax in enumerate(axes[1]):
    ax.hist(X_train_standardized[:, i], bins=25, color='salmon', alpha=0.8, edgecolor='black')
    ax.set_title(titles_std[i])

plt.tight_layout()
plt.show()
```
